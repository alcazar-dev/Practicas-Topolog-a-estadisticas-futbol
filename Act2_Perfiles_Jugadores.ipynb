```python
import pandas as pd
import numpy as np
import kmapper as km
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from kmapper.cover import Cover
from sklearn.cluster import DBSCAN, KMeans

# 0. Carga de datos
# ----------------
# Asume DataFrame original en 'df'
# df = pd.read_csv("fifa_data.csv")

# 1. Selección y escalado de características
# ------------------------------------------
features = [
    'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic',
    'value_eur', 'wage_eur', 'overall', 'potential'
]
df_clean = df[features].dropna()
X_scaled = StandardScaler().fit_transform(df_clean.values)

# 2. Definición de lentes (lenses)
# --------------------------------
pca = PCA(n_components=2)
lens_pca = pca.fit_transform(X_scaled)
lens_price = StandardScaler().fit_transform(
    df_clean['value_eur'].values.reshape(-1,1)
)
# Unión de lentes (PCA 2D + precio)
lens = np.hstack([lens_pca, lens_price])

# 3. Construcción del grafo topológico con más nodos
# --------------------------------------------------
# Para más nodos, aumentamos significativamente el número de cubos y reducimos overlap\ n_mapper = km.KeplerMapper(verbose=1)
cover = Cover(n_cubes=[12,12,12], perc_overlap=0.5)
graph = mapper.map(
    lens,
    X_scaled,
    cover=cover,
    clusterer=DBSCAN(eps=0.4, min_samples=5)
)

# 4. Clustering global con KMeans para más clusters
# -------------------------------------------------
n_clusters = 8
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
global_labels = kmeans.fit_predict(X_scaled)

# 5. Estadísticas de nodos y etiquetado de perfiles
# ------------------------------------------------
clusters = []
for node, idxs in graph['nodes'].items():
    df_node = df_clean.iloc[idxs]
    clusters.append({
        'node': node,
        'size': len(idxs),
        'mean_overall': df_node['overall'].mean(),
        'mean_value_eur': df_node['value_eur'].mean(),
        'mean_potential': df_node['potential'].mean()
    })
clusters_df = pd.DataFrame(clusters)
# Umbrales relajados basados en percentiles 30/70
medians = clusters_df[['mean_overall','mean_value_eur','mean_potential']].quantile([0.3, 0.7]).loc[0.3]

# 6. Asignación de colores y tooltips según perfil de nodo
# -----------------------------------------------------------
label_to_code = {'Perfil intermedio':0,'Alto, caro y consolidado':1,'Jóvenes baratos con alto potencial':2}
custom_colorscale = [[0.0,'#bbbbbb'],[0.5,'#e6550d'],[1.0,'#31a354']]

tooltips = np.empty(X_scaled.shape[0], dtype=object)
color_vals = np.zeros(X_scaled.shape[0], dtype=int)
for node, idxs in graph['nodes'].items():
    row = clusters_df.loc[clusters_df['node']==node].iloc[0]
    if (row['mean_potential'] >= medians['mean_potential']) and (row['mean_value_eur'] <= medians['mean_value_eur'] * 1.25):
        label = 'Jóvenes baratos con alto potencial'
    elif (row['mean_overall'] >= medians['mean_overall']) and (row['mean_value_eur'] >= medians['mean_value_eur']):
        label = 'Alto, caro y consolidado'
    else:
        label = 'Perfil intermedio'
    code = label_to_code[label]
    for i in idxs:
        color_vals[i] = code
        tooltips[i] = label

# 7. Visualización interactiva con perfiles coloreados
# --------------------------------------------------
mapper.visualize(
    graph,
    path_html="fifa_mapper_more_nodes_clusters.html",
    title="FIFA TDA: Más Nodos y Clusters",
    color_values=color_vals,
    color_function_name="Perfil Jugador",
    colorscale=custom_colorscale,
    custom_tooltips=tooltips,
    include_searchbar=True
)

# 8. Análisis de atributos técnicos por cluster KMeans
# ------------------------------------------------
tech_feats = ['pace','shooting','passing','dribbling','defending','physic']
df_summary = df_clean.copy()
df_summary['cluster'] = global_labels
cluster_attrs = (
    df_summary
    .groupby('cluster')
    .agg({**{feat:'mean' for feat in tech_feats},
          'overall':'mean','value_eur':'mean','potential':'mean','cluster':'size'})
    .rename(columns={'cluster':'count'})
    .reset_index()
)
# Etiquetado de clusters globales
medians_k = cluster_attrs[['overall','value_eur','potential']].quantile([0.3,0.7]).loc[0.3]
def label_cluster(row):
    if (row['potential'] >= medians_k['potential']) and (row['value_eur'] <= medians_k['value_eur'] * 1.25):
        return 'Jóvenes baratos con alto potencial'
    elif (row['overall'] >= medians_k['overall']) and (row['value_eur'] >= medians_k['value_eur']):
        return 'Alto, caro y consolidado'
    else:
        return 'Perfil intermedio'
cluster_attrs['profile'] = cluster_attrs.apply(label_cluster, axis=1)

# Mostrar resumen en consola
print("Atributos técnicos medios y perfil por cluster KMeans:")
print(cluster_attrs)
# cluster_attrs.to_csv("cluster_attrs_with_profile.csv", index=False)
```

```python
# Construcción de dataframe con métricas por nodo TDA
tda_clusters = []
for node, idxs in graph['nodes'].items():
    df_node = df_clean.iloc[idxs]
    tda_clusters.append({
        'node': node,
        'size': len(idxs),
        'mean_overall': df_node['overall'].mean(),
        'mean_value_eur': df_node['value_eur'].mean(),
        'mean_potential': df_node['potential'].mean(),
        'mean_pace': df_node['pace'].mean(),
        'mean_shooting': df_node['shooting'].mean(),
        'mean_passing': df_node['passing'].mean(),
        'mean_dribbling': df_node['dribbling'].mean(),
        'mean_defending': df_node['defending'].mean(),
        'mean_physic': df_node['physic'].mean()
    })
tda_df = pd.DataFrame(tda_clusters)

# Cálculo de percentiles 30 para etiquetado
percentiles_tda = tda_df[['mean_overall','mean_value_eur','mean_potential']].quantile([0.3, 0.7]).loc[0.3]

def label_tda_node(row):
    if (row['mean_potential'] >= percentiles_tda['mean_potential']) and (row['mean_value_eur'] <= percentiles_tda['mean_value_eur'] * 1.25):
        return 'Jóvenes baratos con alto potencial'
    elif (row['mean_overall'] >= percentiles_tda['mean_overall']) and (row['mean_value_eur'] >= percentiles_tda['mean_value_eur']):
        return 'Alto, caro y consolidado'
    else:
        return 'Perfil intermedio'

# Asignar perfil a cada nodo
tda_df['profile'] = tda_df.apply(label_tda_node, axis=1)

# Mostrar resumen de los 10 nodos más grandes con su perfil
print("Resumen de los 10 nodos TDA más grandes con perfil asignado:")
print(
    tda_df
    .sort_values(by="size", ascending=False)
    .head(10)
    [['node', 'size', 'mean_overall', 'mean_value_eur', 'mean_potential', 'profile']]
)
```
